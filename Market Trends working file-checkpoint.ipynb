```python
import os
import yaml
import pandas as pd

input_folder = r"E:\Market Trends Visualization\data_raw"
output_folder = "E:\Market Trends Visualization\data_clean_Final"

os.makedirs(output_folder, exist_ok=True)

# Dictionary to store all data for each stock
stock_data = {}

for root, dirs, files in os.walk(input_folder):
    for file in files:
        if file.endswith(".yaml"):
            file_path = os.path.join(root, file)

            with open(file_path, "r") as f:
                data = yaml.safe_load(f)

            df = pd.DataFrame(data)
            

            # Extract stock name
            symbol = df["Ticker"].iloc[0]
          
            # Add to dictionary
            if symbol not in stock_data:
                stock_data[symbol] = df
            else:
                stock_data[symbol] = pd.concat([stock_data[symbol], df])
                

# Save final CSVs
for symbol, df in stock_data.items():
    df = df.sort_values("date")  # sort by date
    df.to_csv("E:/Market Trends Visualization/data_clean_final/cleaned_data.csv", index=False)


print("All YAML files merged and saved successfully!")
```

```python
df.head()
df = df.rename(columns={
    'Tiker': 'Ticker',
    'date': 'Date',
    'open': 'Open',
    'high': 'High',
    'low': 'Low',
    'close': 'Close',
    'volume': 'Volume',
    'month': 'Month'
})
df.isnull().sum()
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by=['Ticker', 'Date'])
df['Daily_Return'] = df.groupby('Ticker')['Close'].pct_change()
df['Cumulative_Return'] = df.groupby('Ticker')['Daily_Return'].transform(
    lambda x: (1 + x).cumprod() - 1
)

# Montly return calculation
# Step 1: Extract month
df['Month'] = df['Date'].dt.month

# Step 2: First and last close per (Ticker, Month)
monthly = df.groupby(['Ticker', 'Month'])['Close'].agg(['first', 'last'])

# Step 3: Monthly return formula
monthly['Monthly_Return'] = (monthly['last'] - monthly['first']) / monthly['first']

# Step 4: Merge back into main DataFrame
df = df.merge(monthly['Monthly_Return'], on=['Ticker', 'Month'], how='left')

volatility = df.groupby('Ticker')['Daily_Return'].std().reset_index()
volatility = volatility.rename(columns={'Daily_Return': 'Volatility'})




```

```python
#Top 5 Gainers and Losers (Month-wise):
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month

monthly_returns = (
    df.groupby(["Year", "Month", "Ticker"])["Close"]
      .agg(["first", "last"])
      .reset_index()
)

monthly_returns["Monthly_Return"] = (monthly_returns["last"] / monthly_returns["first"]) - 1

#Identify Top 5 Gainers & Losers for Each Month
monthly_gainers = {}
monthly_losers = {}

for (year, month), group in monthly_returns.groupby(["Year", "Month"]):
    top5 = group.sort_values("Monthly_Return", ascending=False).head(5)
    bottom5 = group.sort_values("Monthly_Return").head(5)
    
    monthly_gainers[(year, month)] = top5
    monthly_losers[(year, month)] = bottom5


#Dashboard‑Style Visualization (12 Months)
import matplotlib.pyplot as plt
import seaborn as sns

# Get all (year, month) pairs that exist in the dataset
month_year_pairs = sorted(monthly_returns[["Year", "Month"]].drop_duplicates().values.tolist())

plt.figure(figsize=(18, 40))
plot_no = 1

for year, m in month_year_pairs:

    g = monthly_gainers.get((year, m))
    l = monthly_losers.get((year, m))

    # Skip if month has no data
    if g is None or l is None:
        continue

    # Top 5 Gainers
    plt.subplot(len(month_year_pairs), 2, plot_no)
    sns.barplot(
        x="Monthly_Return",
        y="Ticker",
        data=g,
        hue="Ticker",
        palette="Greens_r",
        legend=False
    )
    plt.title(f"Top 5 Gainers - {year}-{m}")
    plot_no += 1

    # Top 5 Losers
    plt.subplot(len(month_year_pairs), 2, plot_no)
    sns.barplot(
        x="Monthly_Return",
        y="Ticker",
        data=l,
        hue="Ticker",
        palette="Reds_r",
        legend=False
    )
    plt.title(f"Top 5 Losers - {year}-{m}")
    plot_no += 1

plt.tight_layout()
plt.show()
```
